```python
import pandas as pd
```

```python
autos = pd.read_csv('autos.csv',encoding = 'Latin-1')
autos.info()
autos.head()
```

```python
autos.columns
```

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'kilometer', 'registration_month', 'fuel_type', 'brand',
       'notepaired_damage', 'date_created', 'num_pictures', 'postal_code',
       'last_seen']
autos.head()
```

```python
autos.describe(include='all')
```

```python
autos["num_pictures"].value_counts()
```

```python
autos = autos.drop(["num_pictures","seller","offer_type"], axis=1)
```

```python
autos.columns
```

```python
autos["kilometer"].value_counts()
```

```python
print(autos["price"].describe())
autos["price"].value_counts().head(20)
```

```python
autos["price"].value_counts().sort_index(ascending=False).head(20)
```

```python
autos["price"].value_counts().sort_index(ascending=True).head(20)
```

```python
autos = autos[autos["price"].between(1,351000)]
autos["price"].describe()
```

```python
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

```python
autos["brand"].value_counts(normalize=True)
```

```python
brand_counts = autos["brand"].value_counts(normalize=True)
common_brands = brand_counts[brand_counts > 0.05].index
print(common_brands)
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

```python
bmp_series = pd.Series(brand_mean_prices)
auto_df = pd.DataFrame(bmp_series,columns = ["mean_price"])
```

```python
auto_df.plot()
```

```python
brand_mean_mileage = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_mileage = brand_only["kilometer"].mean()
    brand_mean_mileage[brand] = int(mean_mileage)

mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending = False)
mean_prices = pd.Series(brand_mean_mileage).sort_values(ascending = False)

brand_info = pd.DataFrame (mean_mileage,columns=['mean_mileage'])
brand_info
```

```python
brand_info.plot(kind="bar")
```
